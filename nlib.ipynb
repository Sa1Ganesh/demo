```python
# Stack using list
stack = []

# Push
stack.append(10)
stack.append(20)
stack.append(30)
print("Stack after pushes:", stack)

# Pop
print("Popped element:", stack.pop())
print("Stack after pop:", stack)

# Peek
print("Top element:", stack[-1])

# isEmpty
print("Is stack empty?", len(stack) == 0)
```

```python
x=np.array([23,24,25,26])
y1=np.array([10,20,30,40])
y2=np.array([5,10,25,20])
linestyle=dict(marker=".",
markersize=15,
markerfacecolor="green",
markeredgecolor="red",linestyle="solid"
,linewidth=4)
plt.plot(x,y1,color="blue",**linestyle
)
plt.plot(x,y2,color="red",**linestyle)
plt.title("Title",fontsize=15,family="Arial",fontweight="bold",color="orange")
plt.xlabel("x axis",fontsize=10,fontweight="bold",color="black")
plt.ylabel("y axis",color="green")
plt.xticks(x)
plt.tick_params(axis="both",color="red")
# plt.yticks()
plt.grid(axis="y",linewidth=2,color="lightgray",linestyle="dashed")
plt.show()
```
